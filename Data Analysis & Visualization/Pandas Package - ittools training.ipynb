## Data structure - "Series"

### pandas.Series(data, index, dtype, copy)
~ we can pass values from numpy arrays, lists, sictionaries, etc.

~ Series is one-dimensional

```python
# Create empty series

import pandas as pd
s = pd.Series()
print(s)
```

```python
# Create Series using Numpy Array

import pandas as pd
import numpy as np

d = np.array(['Fiat', 'Ford', 'Honda', 'Hyundai'])

s = pd.Series(d)
print(s)
```

```python
# Pass Indexing in Series data structure

import pandas as pd
import numpy as np

d = np.array(['Fiat', 'Ford', 'Honda', 'Hyundai'])

s = pd.Series(d, index=[1001,1002,1003,1004])
print(s)
```

```python
# Create Series using dictionaries

import pandas as pd
import numpy as np

dictionary = {'X1':1, 'X2':2, 'X3':3, 'X4':4}

s = pd.Series(dictionary, index=['X1', 'X2', 'X3', 'X4', 'X5'])
print(s)

# NaN (Not a Number)
```

## Data structure - "Data Frame"

### pandas.DataFrame(data, index, columns, dtype, copy)
~ It's a two-dimensional data structure

~ Size Mutable

~ Perform airthmatic operations on data structure

~ Rows and Columns are labeled in data frame data structure

```python
# Create empty Data Frame

import pandas as pd

d = pd.DataFrame()
print(d)
```

```python
# Store values in DataFrame from list

import pandas as pd

l = [['Sean', 2000], ['Harvey', 3000], ['Nick', 1200], ['Kara', 1700]]

d = pd.DataFrame(l)
print(d)
```

```python
import pandas as pd

l = [['Sean', 2000], ['Harvey', 3000], ['Nick', 1200], ['Kara', 1700]]

d = pd.DataFrame(l, columns=['Salesman', 'Sales'], dtype=float)
print(d)
```

```python
# Store values from dictionaries

dictionary = {'Salesman': ['Sean', 'Harvey', 'Nick', 'Kara'], 'Sales':[2000,3000,1200,1700]}

d = pd.DataFrame(dictionary)
print(d)

# No need to take care about columns in the case of dictionaries
```

## Data structure - "Panel"

### pandas.Panel(data, items, major_axis, Minor_axis, dtype, copy)

~ It's a 3-D container

~ items, axis = 0 : correspondse to dataframe container

~ major_axis = 1 : correspondse to the raw index for all dataframes

~ Minor_axis = 2 : columns of the dataframe

```python
# Using dictionaries

import pandas as pd

df1 = pd.DataFrame({'Salesman':['Sean', 'Harvey', 'Mike', 'Kara'], 'Sales':[1200, 2000, 2500, 1700]})

data = {'item1':df1, 'item2':df1}

panel = pd.Panel.from_dict(data, orient='minor')

print(panel)
```

## Dealing with missing values

```python
import pandas as pd
import numpy as np

dictionary = {'Salesman':['Sean', 'Kara', 'Mike', 'Harvey'], 'Sales':[1200, np.nan, 1700, np.nan]}

df2 = pd.DataFrame(dictionary)
df2
```

## isnull()

```python
df2.isnull()
```

## Filter null values

```python
output = pd.isnull(df2['Sales'])
df2[output]
```

## notnull()

```python
df2.notnull()
```

## Filter notnull vlaues

```python
output = pd.notnull(df2['Sales'])
df2[output]
```

## Deal with missing value - fillna()

```python
import pandas as pd
import numpy as np

dictionary = {'Salesman':['Sean', 'Kara', 'Mike', 'Harvey'], 'Sales':[1200, np.nan, 1700, np.nan]}

df3 = pd.DataFrame(dictionary)
print(df3)

df3.fillna(0, inplace = True)
df3
```

```python
df3
# Permanently replaced
```

## Deal with missing values - replace()

```python
print(df2)

df2.replace(to_replace = np.nan, value = 0)
```
